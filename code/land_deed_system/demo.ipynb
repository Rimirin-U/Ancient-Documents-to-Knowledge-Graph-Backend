# Ancient Land Deed Analysis System Demo

This notebook demonstrates the capabilities of the Ancient Land Deed Analysis System.

```python
import requests
import json

BASE_URL = "http://localhost:8000/api/v1"
```

## 1. Parse a Land Deed
Extract entities and relationships from raw text.

```python
sample_text = "立杜绝卖契人张三，今因正用不足，情愿将自己祖遗坐落土名南乡一都二图，地名大树下，田三亩六分，东至李四田，西至沟，南至路，北至王五山，四至分明。凭中立契，出卖与赵六名为业。三面言定，时值价纹银五十两整。其银笔下交足，并无短少。其田自卖之后，任凭赵六管业，张三日后子孙不得言赎，亦不得言找。恐口无凭，立此杜绝卖契为照。嘉庆五年三月十二日。凭中人李四、王五。代书人陈七。"

response = requests.post(f"{BASE_URL}/parse", json={"text": sample_text})
if response.status_code == 200:
    print(json.dumps(response.json(), indent=2, ensure_ascii=False))
else:
    print("Error:", response.text)
```

## 2. Normalize Date
Convert Qing Dynasty year to Gregorian date.

```python
date_str = "嘉庆五年三月十二日"
response = requests.post(f"{BASE_URL}/normalize/date", json={"date_str": date_str})
print(response.json())
```

## 3. Normalize Currency
Convert ancient silver value to 2023 RMB estimate.

```python
amount_str = "纹银五十两"
response = requests.post(f"{BASE_URL}/normalize/currency", json={"amount_str": amount_str})
print(response.json())
```

## 4. Translate Text
Translate ancient text to modern Chinese.

```python
response = requests.post(f"{BASE_URL}/translate", json={"text": sample_text})
print(response.json()["translation"])
```

## 5. Logic Correction
Detect conflicts (e.g., Irrevocable Sale vs Redemption).

```python
conflict_text = "立杜绝卖契人张三...日后子孙准许回赎。"
response = requests.post(f"{BASE_URL}/correct", json={"text": conflict_text})
print(json.dumps(response.json(), indent=2, ensure_ascii=False))
```
